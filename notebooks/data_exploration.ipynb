```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/train.csv")
df.head()
```

```python
len(df)
```

```python
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
dev_df   = df_shuffled.iloc[:2000]
test_df  = df_shuffled.iloc[2000:4000]
train_df = df_shuffled.iloc[4000:]
```

```python
train_df.to_csv("../data/preprocessed/train.csv", index=False)
dev_df.to_csv("../data/preprocessed/dev.csv", index=False)
test_df.to_csv("../data/preprocessed/test.csv", index=False)
```
