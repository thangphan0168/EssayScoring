```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/train.csv")
df.head()
```

```python
len(df)
```

```python
def clean_text(text):
    text = text.replace(u'\xa0', u' ').strip()
    return text
```

```python
df["full_text"] = df["full_text"].apply(clean_text)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google/gemma-3-270m")


def get_text_length(text):
    encoding = tokenizer(text, truncation=False)
    return len(encoding["input_ids"])
```

```python
df["length"] = df["full_text"].apply(get_text_length)
```

```python
print("Essays over 1024 tokens:", (df['length'] > 1024).sum())
```

```python
df[df['length'] > 1500]
```

```python
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
dev_df   = df_shuffled.iloc[:2000]
test_df  = df_shuffled.iloc[2000:4000]
train_df = df_shuffled.iloc[4000:]
```

```python
train_df.to_csv("../data/preprocessed/train.csv", index=False)
dev_df.to_csv("../data/preprocessed/dev.csv", index=False)
test_df.to_csv("../data/preprocessed/test.csv", index=False)
```
